# Load

```python
import numpy as np
import pandas as pd
import polars as pl
import os
from pathlib import Path
import sys

import matplotlib.pyplot as plt
import seaborn as sns

dir_raw = str('/Users/brentondickerson/Downloads/breathe-main' + "/data_raw/")
dir_processed = str('/Users/brentondickerson/Downloads/breathe-main' + "/data/")
```

# Functions

```python
def read_data(file_name, dir):
    if dir == "raw":
        data = pl.from_pandas(pd.read_csv(dir_raw+file_name))
    else: 
        data = pl.from_pandas(pd.read_csv(dir_processed+file_name))
    return data

def save_data(data, file_name):
    data.write_csv(dir_processed + file_name + ".csv")
    print(f"{file_name} has been saved.")

def get_country_income_data(file_name):
    data = (
        pd.read_csv(
            dir_raw + file_name, 
            sep=';', 
            names=['category', 'year', 'share'],
        )
        .assign(country=f"{file_name.replace('income_','').replace('.csv','')}")
        )
    return data
```

```python
prisons = read_data("Prison_Investment_by_year.csv", "raw")
education = read_data("Educational_Investment_by_year.csv", "raw")
police = read_data("Police_Investment_by_year.csv", "raw")
```

```python
prisons = prisons.with_columns(pl.lit("Prisons").alias("Investment"), pl.col("observation_date").str.head(4).alias("year"))
prisons = prisons.select("Investment", "year", pl.col("Investment_amount").alias("amount"))
prisons
```

```python
education = education.with_columns(pl.lit("Education").alias("Investment"), pl.col("observation_date").str.head(4).alias("year"))
education = education.select("Investment", "year", pl.col("W583RC1A027NBEA").alias("amount"))
education
```

```python
police = police.with_columns(pl.lit("Police").alias("Investment"), pl.col("observation_date").str.head(4).alias("year"))
police = police.select("Investment", "year", pl.col("amount"))
police
```

```python
investments = pl.concat([prisons, education, police])
investments = investments.select(pl.col("year").cast(int), "Investment", "amount")
```

# Median Income

```python
median_income = read_data("Median_Personal_Income_By_Year_US_Census.csv", "raw").select(
    pl.col("observation_date").str.replace("-01-01", "").cast(pl.Int64).alias("year"), pl.col("MEPAINUSA646N").alias("median_income")
)
median_income
```

# Mean Income by Race

```python
income_mean_by_race = pd.read_csv(dir_raw + "income_by_Race_US_census.csv").dropna()
income_mean_by_race['year'] = income_mean_by_race['Race and Hispanic origin of householder\nand year1'].astype(str).str[:4].astype(int)
income_mean_by_race
```

```python
shares_wid = read_data("shares_wid", "processed")
shares_wid
```

```python
income_mean_by_race_buckets = pl.from_pandas(income_mean_by_race).filter(
    pl.col('RACE').str.contains("COMBINATION") == False, pl.col("Mean Income") != 'N').select(
        "year", pl.col("RACE").str.replace(" ALONE", "").alias("race"), pl.col("Under $15,000").alias("under_15k"), pl.col("$25,000 \nto\n $34,999").alias("15_25k"),
        pl.col("$25,000 \nto\n $34,999").alias("25_35k"), pl.col("$35,000 \nto\n $49,999").alias("35_50k"), pl.col('$50,000 \nto\n $74,999').alias("50_75k")
    )

income_mean_by_race_mean_buckets = income_mean_by_race_buckets.with_columns(
            (pl.col("under_15k") + pl.col("15_25k") + pl.col("25_35k") + pl.col("35_50k") + pl.col("50_75k")
            ).alias("percentile_in_bottom_50"))
income_mean_by_race_mean_buckets.filter(pl.col("year") == 2005)
```

```python
income_mean_by_race_formatted = pl.from_pandas(income_mean_by_race).filter(
    pl.col('RACE').str.contains("COMBINATION") == False, pl.col("Mean Income") != 'N').select(
        "year", pl.col("RACE").str.replace(" ALONE", "").alias("race"), pl.col("Mean Income").str.replace(',', '').alias("mean income")
    )

all_race_income = income_mean_by_race_formatted.filter(pl.col("race") == 'ALL RACES').select(
    'year', pl.col("mean income").alias("all races mean income")
)
all_race_average = income_mean_by_race_formatted.join(all_race_income, on=["year"])
all_race_average = all_race_average.cast({"mean income": pl.UInt64, "all races mean income": pl.UInt64})
race_multiplier_df = all_race_average.with_columns(
    race_multiplier = pl.col('mean income') / pl.col('all races mean income')).select(
        'year', 'race', 'race_multiplier'
    )

save_data(race_multiplier_df, "race_multiplier")
race_multiplier_df.filter(pl.col("year") == 2005)
```

```python
shares_wider = shares_wid.filter(pl.col("country") == 'usa').join(race_multiplier_df, on="year")
shares_wider = shares_wider.with_columns(
    bottom_income_with_race_multiplier = pl.col("bottom_income_mean") * pl.col("race_multiplier")
    , top_income_with_race_multiplier = pl.col("top_income_mean") * pl.col("race_multiplier")
    )
shares_wider = shares_wider.select(
    'year', 'race', 'bottom_income_with_race_multiplier', 'top_income_with_race_multiplier'
    )
shares_wider.filter(pl.col("year") == 2005)
```

# Median Income by Race

```python

median_income_by_race = median_income.join(race_multiplier_df, on=["year"])
median_income_by_race = median_income_by_race.with_columns(
    (pl.col("median_income") * pl.col("race_multiplier")).alias("race_median_income")
).select("year", "race", "race_median_income")
median_income_by_race
```

# Median Income by Gender 

```python
gender_multiplier = read_data("income_by_Sex_US_census.csv", "raw").select(pl.col("Year").str.slice(0,4).cast(pl.Int64).alias("year"), pl.col("FEMALE_TO_MALE_EARNINGS_MULTIPLIER").alias("fm_gender_multiplier"))
gender_median_income = median_income.join(gender_multiplier, on=["year"]).with_columns(
    (pl.col("median_income") * pl.col("fm_gender_multiplier")).alias("median_income_female"),
)
female_income = gender_median_income.select(pl.col("year"), pl.lit("FEMALE").alias("gender"), pl.col("median_income_female").alias("median_income"))
male_income = gender_median_income.select(pl.col("year"), pl.lit("MALE").alias("gender"), pl.col("median_income"))

gender_median_income_final = pl.concat([male_income,female_income], how="vertical_relaxed")
gender_median_income_final
```

```python
f, ax = plt.subplots(1,1, figsize=[22,8])
sns.lineplot(data=gender_median_income_final, x='year', y='median_income', hue='gender', ax=ax)

last = gender_median_income_final.filter(pl.col('year')==maxyear)
for GENDER in last['gender']:
    data = last.filter(pl.col('gender')==GENDER)
    plt.text(x=data['year'].to_numpy()[0], y=data['median_income'].to_numpy()[0], s=data['gender'].to_numpy()[0])

plt.title("How Does Gender Impact Median income?", fontsize=28)
```
